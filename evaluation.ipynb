```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from the CSV file
data = pd.read_csv('Result/overall_accuracy_comparison_label_flipping.csv')

# Extract datasets and corresponding accuracies
datasets = data['Dataset']
before = data['Before']
without_fuldebug = data['Without']
with_fuldebug = data['With']

# Define the positions for the bars
x = list(range(len(datasets)))
width = 0.25

# Use the Set2 palette (3 colors)
palette = sns.color_palette("Set2", 3)
color_before = palette[0]
color_without = palette[1]
color_with = palette[2]

# Create a figure with an appropriate size
plt.figure(figsize=(10, 6))

# Plot the bars
plt.bar([p - width for p in x], before, width=width, label='Before Unlearn', color=color_before)
plt.bar(x, without_fuldebug, width=width, label='Unlearn Without FULDebug', color=color_without)
plt.bar([p + width for p in x], with_fuldebug, width=width, label='Unlearn With FULDebug', color=color_with)

# Add labels and title with larger fonts
plt.xlabel('Dataset', fontweight='bold', fontsize=14)
plt.ylabel('Accuracy (%)', fontweight='bold', fontsize=14)
plt.xticks(x, datasets, fontsize=12)
plt.ylim(30, 85)
plt.grid(True, linestyle='--', alpha=0.7)

# Place legend below the plot
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=3, fontsize=12)

plt.tight_layout()
plt.savefig('Result/overall_accuracy_comparison_label_fliiping.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file. Adjust the path if needed.
df = pd.read_csv('Result/overall_accuracy_all.csv')

# Melt the DataFrame so that each condition becomes a row.
df_long = pd.melt(df, id_vars=['Dataset', 'Method'], 
                  value_vars=['Global_Before', 'Unlearned', 'Unlearned_with_FULDebug'],
                  var_name='Condition', value_name='Accuracy')

# Map condition names to reader-friendly labels.
mapping = {
    'Global_Before': 'Before Unlearn',
    'Unlearned': 'Unlearn Without FULDebug',
    'Unlearned_with_FULDebug': 'Unlearn With FULDebug'
}
df_long['Condition'] = df_long['Condition'].replace(mapping)

# Set a fixed order for the conditions.
condition_order = ['Before Unlearn', 'Unlearn Without FULDebug', 'Unlearn With FULDebug']
df_long['Condition'] = pd.Categorical(df_long['Condition'], categories=condition_order, ordered=True)

# For consistent ordering on the x-axis, define the order for the Dataset column.
dataset_order = ['MNIST', 'Fashion-MNIST', 'CIFAR10']
df_long['Dataset'] = pd.Categorical(df_long['Dataset'], categories=dataset_order, ordered=True)

# Set the overall style
sns.set(style="whitegrid", font_scale=1.1)

# Create a facet grid: one subplot per algorithm (Method).
g = sns.catplot(
    data=df_long, 
    kind='bar', 
    x='Dataset', 
    y='Accuracy', 
    hue='Condition',
    col='Method', 
    col_wrap=3, 
    height=5, 
    aspect=1, 
    palette='Set2', 
    order=dataset_order,
    legend_out=True
)

# Set titles and axis labels.
g.set_titles("{col_name}", fontweight='bold')
g.set_axis_labels("Dataset", "Accuracy (%)", fontweight='bold')
g._legend.set_title("Condition")

# Remove the default legend and add a common one at the bottom.
g._legend.remove()
g.fig.legend(loc='lower center', bbox_to_anchor=(0.5, -0.08), ncol=3, fontsize=12, title_fontsize=12)

# plt.suptitle("Overall Accuracy Comparison for Each Algorithm", fontweight='bold', fontsize=18, y=1.05)
plt.tight_layout()
plt.savefig('Result/overall_accuracy_all.png', dpi=300, bbox_inches='tight')
plt.show()
```
