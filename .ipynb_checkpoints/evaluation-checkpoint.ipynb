```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file
df_loaded = pd.read_csv('Result/ibm_ful_vs_feddebug.csv')

# Extracting data for plotting
time_points = df_loaded['Time (seconds)']
ibm_ful_times = df_loaded['IBM-FUL']
ibm_ful_with_feddebug_times = df_loaded['IBM-FUL with FEDDebug']

# Set bar width and positions
bar_width = 0.3
r1 = np.arange(len(time_points))
r2 = [x + bar_width for x in r1]

# Create the grouped bar plot
plt.figure(figsize=(4, 3))
plt.bar(r1, ibm_ful_times, color='teal', width=bar_width, edgecolor='grey', label='Without FULDebug')
plt.bar(r2, ibm_ful_with_feddebug_times, color='coral', width=bar_width, edgecolor='grey', label='With FULDebug')

# Add labels and title
plt.xlabel('Number of Clients', fontweight='bold')
plt.ylabel('Duration (seconds)', fontweight='bold')
# plt.title('Runtime overhead of the Framework', fontweight='bold')

# Setting the ticks and labels for the x-axis
plt.xticks([r + bar_width/2 for r in range(len(time_points))], time_points)

# Add a legend
plt.legend()

plt.savefig('Result/images/runtime_overhead.png', dpi=300, bbox_inches='tight')
# Show plot
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd

# Load data from CSV files
mnist_df = pd.read_csv('Result/mnist_unlearning_fuldebug.csv')
fashion_mnist_df = pd.read_csv('Result/fashion_mnist_unlearning_fuldebug.csv')
cifar10_df = pd.read_csv('Result/cifar10_unlearning_fuldebug.csv')

# Create subplots (horizontally aligned)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3))

# Plot for MNIST
ax1.plot(mnist_df['Class'], mnist_df['Global Model Before Accuracy (%)'], marker='o', color='blue', label='Before Unlearning', linewidth=1)
ax1.plot(mnist_df['Class'], mnist_df['Global Model After Accuracy Without FULDEBUG (%)'], marker='s', color='red', label='Without FULDebug', linewidth=1)
ax1.plot(mnist_df['Class'], mnist_df['Global Model After Accuracy With FULDEBUG (%)'], marker='^', color='green', label='With FULDebug', linewidth=2)
ax1.set_title('MNIST')
ax1.set_xlabel('Classes', fontweight='bold')
ax1.set_ylabel('Accuracy (%)', fontweight='bold')
ax1.grid(True)

# Plot for Fashion-MNIST
ax2.plot(fashion_mnist_df['Class'], fashion_mnist_df['Global Model Before Accuracy (%)'], marker='o', color='blue', linewidth=1)
ax2.plot(fashion_mnist_df['Class'], fashion_mnist_df['Global Model After Accuracy Without FULDEBUG (%)'], marker='s', color='red', linewidth=1)
ax2.plot(fashion_mnist_df['Class'], fashion_mnist_df['Global Model After Accuracy With FULDEBUG (%)'], marker='^', color='green', linewidth=1.5)
ax2.set_title('Fashion-MNIST')
ax2.set_xlabel('Classes', fontweight='bold')
ax2.grid(True)

# Plot for CIFAR-10
ax3.plot(cifar10_df['Class'], cifar10_df['Global Model Before Accuracy (%)'], marker='o', color='blue', linewidth=1)
ax3.plot(cifar10_df['Class'], cifar10_df['Global Model After Accuracy Without FULDEBUG (%)'], marker='s', color='red', linewidth=1)
ax3.plot(cifar10_df['Class'], cifar10_df['Global Model After Accuracy With FULDEBUG (%)'], marker='^', color='green', linewidth=1.5)
ax3.set_title('CIFAR-10')
ax3.set_xlabel('Classes', fontweight='bold')
ax3.grid(True)

# Common legend at the bottom of the plot
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.1))

# Adjust layout to fit all plots
plt.tight_layout()
plt.savefig('Result/images/class_wise_analysis.png', dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
data = pd.read_csv('Result/overall_accuracy_comparison.csv')

# Extract datasets and corresponding accuracies
datasets = data['Dataset']
before = data['Before']
without_fuldebug = data['Without']
with_fuldebug = data['With']

# Define the positions for the bars
x = range(len(datasets))
width = 0.25

# Create a bar plot
plt.figure(figsize=(5, 3))
plt.bar([p - width for p in x], before, width=width, label='Before', color='goldenrod')
plt.bar([p + width for p in x], with_fuldebug, width=width, label='With FULDEBUG', color='coral')
plt.bar(x, without_fuldebug, width=width, label='Without FULDEBUG', color='Teal')

# Add labels and title
plt.xlabel('Dataset', fontweight='bold')
plt.ylabel('Accuracy (%)', fontweight='bold')
# plt.title('Overall Accuracy')
plt.xticks(x, datasets)
plt.ylim(60, 100)
plt.grid(True)
plt.legend()
# Add legend below the plot
# plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)
plt.savefig('Result/images/overall_accuracy_comparison.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.tight_layout()
plt.show()
```
